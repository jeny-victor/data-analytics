7. Can we visualize the correlation between different pre-existing conditions using a heatmap?

8. How does age correlate with the different COVID test findings?


```python
test_findings =['INTUBED','PNEUMONIA','DIABETES','ASTHMA','HIPERTENSION','OBESITY','RENAL_CHRONIC','TOBACCO']
age_test_findings = covid[['AGE'] + test_findings]
age_test_findings.corr()
```

9. What is the distribution of ages among different outcomes (recovered, died)?



```python
def outcome(x):
    if x == 1:
        return 'True'
    else:
        return 'False'
```

```python
covid['Outcomes'] = covid['PATIENT_TYPE'].apply(outcome)
covid.groupby('Outcomes')['AGE'].hist(alpha =0.5,bins=20)
covid[covid['Outcomes'] =='Recovered ']['AGE'].hist(alpha =0.5,bins= 20,label ='Recovered')
covid[covid['Outcomes']=='Died']['AGE'].hist(alpha =0.5,bins = 20,color = 'blue',label ='Died')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
